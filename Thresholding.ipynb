# Thresholding operations and plots

```python
from IPython.core.pylabtools import figsize
figsize(8, 8)

from munge.DataElement import DataElement
from munge.ImageThresholder import ImageThresholder
```

### Define a DataElement with DCM and contour files

```python
element = DataElement('data/dicoms/SCD0000101/59.dcm',
                      'data/contourfiles/SC-HF-I-1/i-contours/IM-0001-0059-icontour-manual.txt',
                      'data/contourfiles/SC-HF-I-1/o-contours/IM-0001-0059-ocontour-manual.txt')
```

### Define a ImageThresholder

```python
thresholder = ImageThresholder(element, n_components=2, method='gmm', postprocess=True)
```

### Plot the model fit of the thresholder

```python
thresholder.plot_model_fit()
```

### Tri-modal fit
